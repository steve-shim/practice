The Figure is the top-level container in this hierarchy. It is the overall window/page that everything is drawn on. You can have multiple independent figures and Figures can contain multiple Axes.

Most plotting ocurs on an Axes. The axes is effectively the area that we plot data on and any ticks/labels/etc associated with it. Usually we'll set up an Axes with a call to subplot (which places Axes on a regular grid), so in most cases, Axes and Subplot are synonymous.

Each Axes has an XAxis and a YAxis. These contain the ticks, tick locations, labels, etc. In this tutorial, we'll mostly control ticks, tick labels, and data limits through other mechanisms, so we won't touch the individual Axis part of things all that much. However, it is worth mentioning here to explain where the term Axes comes from.

figure 생성

```python
%matplotlib inline 
import numpy as np
import matplotlib.pyplot as plt

fig = plt.figure() # plt.figure()는 새로운 figure를 생성해준다
plt.show()         # plt.show()는 생성된 모든 figure를 보여준다
```

```python
fig = plt.figure(figsize=(9,9)) # 최초 창의 크기를 가로 세로 9인치로 설정
```

axes 생성  
axes는 figure 내에서 축을 가지는 하나의 좌표평면과 같은 개념이다. 실제로 데이터가 그려지는 곳은 axes이다.
add_axes() 메서드를 쓰면 해당 figure 클래스에 axes를 생성할 수 있지만, 여기서는 add_subplot()을 써보자.

subplot은 하나의 figure에 마치 표처럼 여러 axes를 표현할 수 있게 해준다. 2번째 줄에서 add_subplot(1,1,1)이라고 한 것은 전체 창을 가로 1칸, 세로 1칸으로 쪼개고 그 중 첫 번째 칸에 ax라는 이름의 axes를 생성한다.

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
plt.show()
```

add_subplot(2,1,1) -> 전체 창을 가로 1칸, 세로 2칸으로 쪼개고, 그 중 첫 번째 칸에 ax1, 두 번째 칸에 ax2를 생성한다.

```python
fig = plt.figure()
ax1 = fig.add_subplot(1,2,1)
ax2 = fig.add_subplot(1,2,2)
plt.show()
```

axes 설정
axes마다 형식을 지정해 줄 수 있다.

```python
fig = plt.figure()
ax1 = fig.add_subplot(1,2,1)
ax2 = fig.add_subplot(1,2,2)
 
ax1.set(xlim=[0., 7.], ylim=[-0.5, 40.5], title='Example', xlabel='xAxis', ylabel='yAxis')

ax2.set_xlim([0., 20.])
ax2.set_ylim([-1.5, 50.5])
ax2.set_title('Example2', size=20)
ax2.set_xlabel('xAxis', size=10)
ax2.set_ylabel('yAxis', size=15)

x = np.array([1,3,6,4])
y = np.array([10,20,25,40])
ax1.plot(x, y)
ax2.plot(x, y, color='red', linewidth=3)
x = np.array([0.3, 3.8, 1.2, 2.5])
y = np.array([11, 25, 9, 26])
ax1.scatter(x, y, marker='^', color='darkgreen')

plt.tight_layout()
plt.show()
```
